1. Find all songs(tracks) by Pink Floyd

```python
%sql SELECT T."Name", AR."Name" FROM "Track" T\
LEFT JOIN "Album" A on T."AlbumId" = A."AlbumId"\
LEFT JOIN "Artist" AR on A."ArtistId" = AR."ArtistId"\
WHERE AR."Name" = 'Pink Floyd'
```

2. Find all Rock songs performed by any artist whose name starts with a 'D'


```python
%sql SELECT T."Name", AR."Name", G."Name" From "Track" T\
LEFT JOIN "Genre" G on T."GenreId" = G."GenreId"\
LEFT JOIN "Album" A on T."AlbumId" = A."AlbumId"\
LEFT JOIN "Artist" AR on A."ArtistId" = AR."ArtistId"\
WHERE AR."Name" like 'D%' and G."Name" = 'Rock'
```

3. What genres of songs are performed in the album called 'Morning Dance'?


```python
%sql SELECT G."Name" from "Genre" G WHERE G."GenreId" = (SELECT DISTINCT T."GenreId" from "Album" A, "Track" T\
WHERE A."AlbumId" = T."AlbumId" and A."Title" = 'Morning Dance')
```

4. Which artists have performed in songs of more than one genre?


```python
%sql SELECT AR."Name" as "Artist_Name", count(distinct T."GenreId") as "Number" from "Track" T\
LEFT JOIN "Album" A on T."AlbumId" = A."AlbumId"\
LEFT JOIN "Artist" AR on A."ArtistId" = AR."ArtistId"\
GROUP BY AR."Name" HAVING count(distinct T."GenreId") > 1
```

5. Which composers have written songs for more than one album?


```python
%sql WITH ctable AS (SELECT A."AlbumId",regexp_split_to_table("Composer",E'[/,&-]') as composer from "Track" T\
LEFT JOIN "Album" A on A."AlbumId"=T."AlbumId"\
LEFT JOIN "Artist" AR on AR."ArtistId"=A."ArtistId")\
SELECT RTRIM(LTRIM("composer")) as composer_name,count(distinct "AlbumId") from "ctable"\
GROUP BY "composer_name" HAVING count(distinct "AlbumId") > 1
```
